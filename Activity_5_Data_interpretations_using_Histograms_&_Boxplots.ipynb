```python
def custom_cs(x):
  if (0 <=x<35):
    return "<35"
  if(35 <= x<=54):
      return "35 - 54"
  if (55 <= x <74):
        return "55 - 74"
  if (75 <= x <94):
        return "75 - 94"
  if (95 <= x <115):
        return "95 - 115"
  return ">=115"
```

```python
df['ag']= df.apply(lambda tbl: custom_cs(tbl['peso']),
                      axis = 1)
df['ag']
```

```python
ag = df.groupby(['peso']).size()
ag = ag.to_frame()
ag
```
