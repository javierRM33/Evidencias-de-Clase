```python
Path = '/content/drive/MyDrive/ACT3TD/adult.csv'
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(Path)
df
```

```python
df.columns
```

```python
EstadoCivil=pd.pivot_table(df,
               index = ['marital-status'],
               values='age', 
               aggfunc='count')
EstadoCivil.rename(columns={'age':'total'}, inplace=True)
EstadoCivil
```

```python
total = df['age'].count()
EstadoCivil['porcentaje'] = (EstadoCivil['total'] / total * 100).round(3).astype(str) + '%'
EstadoCivil
```

```python
A=pd.pivot_table(df,
               index = ['hours-per-week', 'education'],
                 values='marital-status', 
               )
           
A
```

```python
never_married_bachelors = df[(df['marital-status'] == 'Never-married') & (df['education'] == 'Bachelors')]

# Calcular el promedio de la columna "hours-per-week"
average_hours_per_week = never_married_bachelors['hours-per-week'].mean()

# Imprimir el resultado
print('El promedio de horas por semana de los individuos con estado civil "Never-married" y nivel de educación "Bachelors" es:', average_hours_per_week)
```
