```python
Path = '/content/drive/MyDrive/ACT3TD/antropometria-dataset-2-2.csv'
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import plotly
import plotly.express as px
```

```python
df = pd.read_csv(Path)
df
```

```python
df.columns
```

```python
df['peso'].plot(kind='hist', bins=15)
```

```python
sns.lmplot(x='diastol', 
           y='peso', 
           data=df,
           fit_reg=False,  # No regression line
           hue='sexo')   # Color by evolution stage

plt.show()
```

```python
df['cintura'].plot(kind='hist', bins=15)
```

```python
sns.lmplot(x='sexo', 
           y='cintura', 
           data=df
          )
plt.title('Scatterplot Wingspan vs CWDistance')
plt.show()
```

```python
sns.boxplot(data=df.loc[:,
                        ['cintura']
                        ]);
```

```python
# Filtrar solo las filas con sexo igual a 1
df_sexo1 = df[df['sexo'] == 1]

# Generar el grÃ¡fico para sexo igual a 1
sns.lmplot(x='sistol', 
           y='diastol', 
           data=df_sexo1,
           fit_reg=False,  # No regression line
           hue='sexo')   # Color by evolution stage

# Mostrar el grÃ¡fico
plt.show()
```

```python
# Filter the DataFrame to include only rows where sexo is 1
df_sexo1 = df[df['sexo'] == 2]

# Generate histogram for "sistol" values where sexo is 1
sns.histplot(x='peso', 
             data=df_sexo1,
             hue='sexo',
             bins=25)

# Show the plot
plt.show()
```
