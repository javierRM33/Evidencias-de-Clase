```python
import pandas as pd
path='https://raw.githubusercontent.com/javierRM33/Data-Mining/main/antropometria-dataset-2.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
pd.pivot_table(df, 
               index='desc_ent', 
               values = 'peso',
               aggfunc = 'mean',
               margins = True
              )
```

```python
pt1= pd.pivot_table(df, 
               index=('desc_ent','sexo'), 
               values = 'peso',
               aggfunc = ('mean'),
               margins = True
              )
pt1
```

```python
S1= [1]
```

```python
FS = df[df.sexo.isin(S1)]
FS
```

```python
pt2= pd.pivot_table(FS, 
               index=('desc_ent','sexo'), 
               values = 'peso',
               aggfunc = ('mean'),
               margins = True
              )
pt2
```

```python
min_S1= pt2.min().min()
min_S1
```

```python
import pandas as pd
path= 'https://raw.githubusercontent.com/javierRM33/Data-Mining/main/adult.csv'
```

```python
df2 = pd.read_csv(path)
df2
```

```python
pt3= pd.pivot_table(df2, 
               index=('marital-status','education'),
               values = 'hours-per-week',
               aggfunc = ('mean'),
               margins = True
              )
pt3
```

```python
pt4= pd.pivot_table(df2, 
               index=('marital-status','education'),
               columns = 'sex',
               values = 'age',
               aggfunc = ('median'),
               margins = True
              )
pt4
```
